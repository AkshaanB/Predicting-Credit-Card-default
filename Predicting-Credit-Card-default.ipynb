```python
balance = df_data['Balance_Limit_V1'].values
```

```python
processed_balance = []
for value in balance:
    if value == '100K':
        processed_balance.append(100000)
    elif value == '200K':
        processed_balance.append(200000)
    elif value == '300K':
        processed_balance.append(300000)
    elif value == '400K':
        processed_balance.append(400000)
    elif value == '500K':
        processed_balance.append(500000)
    elif value == '1M':
        processed_balance.append(1000000)
    elif value == '1.5M':
        processed_balance.append(1500000)
    else:
        processed_balance.append(2500000)
```

```python
df_data.columns = ['Balance_Limit', 'Gender', 'MARITAL_STATUS', 'AGE', 'PAY', 'DUE_AMT', 'PAID_AMT', 'NEXT_MONTH_DEFAULT', 'EDUCATION_STATUS_Graduate', 'EDUCATION_STATUS_High School','EDUCATION_STATUS_Other']
```

```python
x = df_data.drop(['NEXT_MONTH_DEFAULT'], axis=1)
y = df_data['NEXT_MONTH_DEFAULT']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

> **Standardizing the independent variables**

```python
standard = preprocessing.StandardScaler()
```

```python
x_train = standard.fit_transform(x_train)
x_test = standard.fit_transform(x_test)
```

**Building the ANN**

```python
import keras
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense
```

**Input layer and 1st hidden layer**

```python
model.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=10))
```

**2nd hidden layer**

```python
model.add(Dense(output_dim=6, init='uniform', activation='relu'))
```

**3rd hidden layer**

```python
model.add(Dense(output_dim=6, init='uniform', activation='relu'))
```

**Output layer**

```python
model.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=10, nb_epoch=100)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
y_pred = label_encoder.fit_transform(y_pred.astype(str))
```

```python
results_df = pd.DataFrame({'actual':y_test, 
                           'predicted':y_pred})
```

```python
results_crosstab = pd.crosstab(results_df['predicted'], results_df['actual'])
```

```python
results_crosstab
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
